```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Mention the url of the website
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
# request the contents
page=requests.get(url)
```

```python
# to access underlying html, use beautifulsoup object
soup=BeautifulSoup(page.text,'html')
```

```python
# retrieve the table we want, we can do it by mentioning its unique class, or use find_all with indexing
# soup.find_all('table')[0]
table=soup.find('table', class_='wikitable sortable')
```

```python
# find out table headers
world_table=table.find_all('th')
```

```python
# retrieve only column names from the table
title= [title.text.strip() for title in world_table]
```

```python
# from the result we can see that some other values are also coming as header
# so we are limiting the titles
title=title[2:10]
```

```python
# create data frame for the columns we obtained above
data= pd.DataFrame(columns=title)
```

```python
data
```

```python
# now take all the rows from the table
rows=table.find_all('tr')
```

```python
rows
```

```python
rows[2].find_all('td')
```

```python
#loop through each row
for row in rows[2:]:
    # take each value od td in each row
    row_ref=row.find_all('td')
    # save all td values in a list
    individual_data=[val.text.strip() for val in row_ref]
    # for indexing purpose we use length of the data
    length=len(data)
    # save the list of values to the dataframe by mentioning the index no
    data.loc[length]=individual_data
```
